```python
# === FINAL COLAB: EXPORT TO CSV ===
import requests, csv
from datetime import datetime
from google.colab import files

# YOUR DATABASE SECRET (from Firebase Console → Service Accounts → Database Secrets)
DATABASE_SECRET = "PASTE_YOUR_DATABASE_SECRET_HERE"

DB = "https://save-in-a-database-default-rtdb.asia-southeast1.firebasedatabase.app"

print("Fetching logs from Firebase...")
url = f"{DB}/esp32/log.json?auth={DATABASE_SECRET}"
data = requests.get(url).json()

if not data or data is None:
    print("No logs! Toggle relay ON/OFF in Serial Monitor.")
else:
    rows = []
    for ts, state in data.items():
        # Remove +05:30 for parsing
        clean_ts = ts.split('+')[0]
        dt = datetime.fromisoformat(clean_ts)
        rows.append({
            'timestamp': ts,
            'state': state,
            'hour': dt.hour,
            'minute': dt.minute,
            'day_of_week': dt.weekday(),  # 0=Mon
            'is_on': 1 if state == 'ON' else 0
        })

    # Save CSV
    with open('relay_logs.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['timestamp','state','hour','minute','day_of_week','is_on'])
        writer.writeheader()
        writer.writerows(rows)

    print(f"Exported {len(rows)} logs to relay_logs.csv")
    print("First 3 logs:")
    for r in rows[:3]: print(r)

    # Download
    files.download('relay_logs.csv')
```
