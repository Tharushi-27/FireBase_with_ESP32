```python
import requests
import csv
from datetime import datetime

DATABASE_URL = "https://userinput-wifi-default-rtdb.asia-southeast1.firebasedatabase.app"
AUTH_TOKEN = "WgZ4LV2V5oFgWOC886CUozmQx6yNHAfen0P6pd0X"

print("Fetching data...")
url = f"{DATABASE_URL}/esp32/log.json?auth={AUTH_TOKEN}"
data = requests.get(url).json()

if not data:
    print("No logs! Turn relay ON/OFF a few times.")
else:
    rows = []
    for ts, state in data.items():
        dt = datetime.fromisoformat(ts.replace('Z', '+00:00'))
        rows.append({
            'timestamp': ts,
            'state': state,
            'hour': dt.hour,
            'minute': dt.minute,
            'day_of_week': dt.weekday(),
            'is_on': 1 if state == 'ON' else 0
        })

    with open('relay_logs.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['timestamp','state','hour','minute','day_of_week','is_on'])
        writer.writeheader()
        writer.writerows(rows)

    print(f"Exported {len(rows)} logs!")
    print("First 3:")
    for r in rows[:3]: print(r)

    from google.colab import files
    files.download('relay_logs.csv')
```
